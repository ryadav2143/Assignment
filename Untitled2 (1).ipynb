```python
Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
MongoDB over SQL databases?
Ans:
    MongoDB is a popular open-source, document-oriented NoSQL (non-relational) database management system. It stores data in JSON-like documents 
with dynamic schemas, allowing flexibility and ease of handling complex data structures. MongoDB is designed to scale horizontally across multiple
servers, making it suitable for handling large amounts of data and high traffic.

Non-relational databases, or NoSQL databases, are a class of databases that do not follow the traditional relational data model. Unlike SQL
databases (Relational databases), which store data in tables with predefined schemas and enforce relationships between tables, NoSQL databases 
store data in a more flexible and schema-less manner. They are designed to handle unstructured or semi-structured data, making them suitable
for various data types, such as documents, key-value pairs, column-family, and graph data.

Scenarios where MongoDB is preferred over SQL databases:
Flexible Schema: When your data structure is not well-defined or changes frequently, MongoDB's flexible schema allows you to store and modify 
data without the need to modify a predefined table structure, unlike SQL databases.

Scalability: MongoDB is designed for horizontal scalability, making it suitable for applications with high data volumes and traffic. It can
distribute data across multiple nodes, ensuring better performance and handling large-scale applications.

Big Data and Real-time Analytics: For applications that require real-time analytics and processing of massive volumes of data, MongoDB's 
ability to handle complex queries and indexes efficiently can be advantageous.

Unstructured Data: When dealing with semi-structured or unstructured data, like documents, arrays, or nested objects, MongoDB's document-oriented
approach simplifies data storage and retrieval.

Cloud-Native Applications: MongoDB is well-suited for cloud environments due to its horizontal scaling capabilities, allowing applications to 
take advantage of cloud infrastructure efficiently.

Rapid Development: In scenarios where speed and agility are essential, MongoDB's flexible nature allows for faster development cycles and quick 
iteration of data models.

Geospatial Data: MongoDB has built-in geospatial queries, making it an excellent choice for applications that handle location-based data.
```

```python
Q2. State and Explain the features of MongoDB?
Ans:
    Document-Oriented: MongoDB is a document-oriented database, where data is stored in BSON (Binary JSON) documents. Each document can have
a different structure, providing flexibility to handle evolving and unstructured data.

Schema Flexibility: MongoDB has a dynamic schema, meaning documents in a collection can have different fields and data types. This allows 
developers to change the data model without the need to modify the entire database schema, making it easy to adapt to changing requirements.

High Scalability: MongoDB can scale horizontally by distributing data across multiple servers or nodes. This enables it to handle large volumes
of data and high throughput, making it suitable for applications with high growth and demand.

High Availability: MongoDB supports replica sets, which are a group of database nodes that maintain identical data copies. If a primary node fails,
a secondary node can take over, ensuring high availability and fault tolerance.

Automatic Sharding: MongoDB supports automatic sharding, a technique that distributes data across multiple shards (partitions) to achieve horizontal
scaling. It automatically balances data across shards as the data grows.

Indexes: MongoDB allows the creation of indexes on fields in documents to optimize query performance. Indexes improve the speed of data retrieval 
and support complex queries efficiently.

Aggregation Framework: MongoDB provides a powerful aggregation framework that allows users to perform data transformation, filtering, grouping, 
and computation within the database. This is useful for real-time analytics and reporting.

Ad Hoc Queries: MongoDB supports dynamic and ad hoc queries, enabling users to perform queries on documents using a flexible query language 
without needing to predefine specific indexes or structures.

Geospatial Indexes and Queries: MongoDB has built-in support for geospatial data. It can store and query location-based information, making it 
ideal for applications that deal with geospatial data.

Full-Text Search: MongoDB offers full-text search capabilities, allowing users to search for text content efficiently within documents.

Encryption and Security: MongoDB provides features like field-level encryption, SSL/TLS support, access control, and authentication mechanisms 
to ensure data security.

Native JSON Support: MongoDB stores data in JSON-like BSON format, making it easy to work with modern web applications that use JSON extensively.

Change Streams: MongoDB supports change streams, which allow applications to subscribe to real-time data changes in the database, enabling reactive
programming and event-driven architectures.
```

```python
Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB?
Ans:
    In MongoDB, the sort() method is used to sort the results of a query. It allows you to specify one or more fields by which the query results 
should be sorted and the order in which the sorting should be applied (ascending or descending).

The basic syntax of the sort() method in pymongo is as follows:
cursor = collection.find(query_filter).sort(sort_key, sort_order)
Where:
collection: The MongoDB collection you want to query.
query_filter: A dictionary that defines the conditions to filter the documents you want to retrieve. The query_filter is optional and can be omitted if you want to sort all documents in the collection.
sort_key: The field by which the sorting should be applied. This should be a field name in the documents.
sort_order: The sorting order. It can be pymongo.ASCENDING for ascending order (1) or pymongo.DESCENDING for descending order (-1).
You can use the sort() method with find() to sort the documents in a specific order based on one or more fields.

Here's an example to demonstrate sorting in MongoDB:
import pymongo
mongodb_connection_string = "mongodb://username:password@your-mongodb-host:port/"
db_name = "your_database_name"
collection_name = "your_collection_name"

# Connect to MongoDB
client = pymongo.MongoClient(mongodb_connection_string)
db = client[db_name]
collection = db[collection_name]

# Sample data to insert into the collection
sample_data = [
    {"name": "John", "age": 30, "city": "New York"},
    {"name": "Alice", "age": 25, "city": "San Francisco"},
    {"name": "Bob", "age": 35, "city": "Los Angeles"},
    {"name": "David", "age": 28, "city": "Chicago"}
]

collection.insert_many(sample_data)

query_filter = {}

cursor = collection.find(query_filter).sort("age", pymongo.ASCENDING)

print("Sorted by Age (Ascending):")
for document in cursor:
    print(document)
    
cursor = collection.find(query_filter).sort("age", pymongo.DESCENDING)

print("\nSorted by Age (Descending):")
for document in cursor:
    print(document)
client.close()
```

```python
Q7. Explain why delete_one(), delete_many(), and drop() is used?
Ans:
    In MongoDB, delete_one(), delete_many(), and drop() are methods used to remove data from the database.

delete_one():
Purpose: It is used to delete a single document that matches a specified filter.
Syntax: collection.delete_one(filter)
Use case: When you want to remove one specific document from the collection based on certain criteria. It deletes the first document that 
matches the given filter.

delete_many():
Purpose: It is used to delete multiple documents that match a specified filter.
Syntax: collection.delete_many(filter)
Use case: When you want to remove multiple documents that meet a certain condition from the collection. It deletes all the documents that 
match the given filter.

drop():
Purpose: It is used to delete an entire collection from the database.
Syntax: collection.drop()
Use case: When you want to remove the entire collection and all the documents it contains. Use this method with caution as it is a non-recoverable
operation, and all data within the collection will be permanently lost.
```
