```python
Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example?
Ans:
 Probability Mass Function (PMF):
PMF is used for discrete random variables, which take on a countable number of distinct values.
It gives the probability that a discrete random variable equals a specific value.
The PMF is denoted as P(X = x), where X is the random variable and x is a specific value.
The sum of all PMF values over all possible values of the random variable should equal 1.

Example:
Let's consider a fair six-sided die. The random variable X represents the outcome of a single roll of the die. The PMF for
this random variable is as follows:
P(X = 1) = 1/6
P(X = 2) = 1/6
P(X = 3) = 1/6
P(X = 4) = 1/6
P(X = 5) = 1/6
P(X = 6) = 1/6
In this example, the PMF gives the probability of getting each of the six possible outcomes when rolling the die.

Probability Density Function (PDF):
PDF is used for continuous random variables, which can take on an infinite number of values within a range.
It represents the probability distribution of a continuous random variable by specifying the likelihood of the variable
taking on a specific value within an interval.
The PDF is denoted as f(x), where x is a specific value.
Example:
Let's consider a continuous random variable Y representing the time it takes for a computer to complete a task.
The PDF for this random variable might be modeled as follows:
f(y) = k * e^(-0.2y), for y ≥ 0
In this equation, f(y) is the PDF, and k is a constant chosen to ensure that the total area under the PDF curve
equals 1. The PDF describes the likelihood of the computer task taking a particular amount of time (y) within the
specified range.
```

```python
Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
Ans:
The CDF gives you the probability that the random variable X is less than or equal to a given value x.
example to illustrate the concept of a CDF:
Let's consider a continuous random variable Y that represents the height (in inches) of adult males in a population.
Suppose we have collected data and found that the CDF for this random variable is as follows:

F(y) = 0 for y < 60
F(y) = 0.2 for 60 ≤ y < 65
F(y) = 0.6 for 65 ≤ y < 70
F(y) = 0.9 for 70 ≤ y < 75
F(y) = 1 for y ≥ 75

In this example, F(y) gives the cumulative probability of a randomly selected adult male having a height less than or equal
to y. For instance:

F(62) = 0.6, which means there is a 60% chance that an adult male's height is less than or equal to 62 inches.
F(70) = 0.9, indicating a 90% chance that an adult male's height is less than or equal to 70 inches.
Why CDF is used:

Probability Calculation: The CDF simplifies the calculation of probabilities involving a random variable. To find
the probability that X falls within a specific range [a, b], you can use the CDF:
P(a ≤ X ≤ b) = F(b) - F(a)

Comparison: You can compare different random variables or distributions by examining their CDFs. This helps in understanding
the characteristics of these variables, such as central tendency, spread, and shape.

Statistical Inference: CDFs are essential for performing various statistical tests and making inferences about population 
parameters, such as finding percentiles, constructing confidence intervals, and conducting hypothesis tests.

Random Variable Description: CDFs provide a concise and comprehensive description of a random variable's distribution, which
can be crucial for modeling and analysis.
```

```python
Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.
Ans:
Height of Individuals: The heights of a population tend to follow a normal distribution. The mean represents the average height,
and the standard deviation indicates the spread or variability in heights.

IQ Scores: IQ scores are often assumed to be normally distributed, with the mean set at 100 and the standard deviation at 15.
This allows for comparisons and assessments of intelligence.

Measurement Errors: When taking repeated measurements of the same quantity (e.g., in laboratory experiments), the distribution
of measurement errors often approximates a normal distribution, with the mean close to zero.

Financial Returns: Daily or monthly returns of stocks or financial assets are often modeled using a normal distribution. 
The mean represents the average return, and the standard deviation quantifies the volatility or risk associated with the asset.

Test Scores: Scores on standardized tests, such as the SAT or GRE, are typically assumed to follow a normal distribution.
The mean score represents the average performance, and the standard deviation reflects the variability in scores.

How Parameters Shape the Distribution:

Mean (μ): The mean is the central value of the distribution. It determines the location of the peak or center of the bell curve.
If μ increases, the peak shifts to the right, and if μ decreases, it shifts to the left.

Standard Deviation (σ): The standard deviation measures the spread or dispersion of the data. A larger σ results in a wider,
flatter curve, indicating greater variability. A smaller σ leads to a narrower, taller curve, signifying less variability.
```

```python
Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?
Ans:
Bernoulli Distribution:
A discrete probability distribution for a single trial with two outcomes: success (usually denoted as 1)
and failure (usually denoted as 0).
Characterized by a single parameter, p, representing the probability of success.
Example: Modeling the outcome of a single coin flip, where heads is success (1) and tails is failure (0).

Difference Between Bernoulli and Binomial Distribution:
Bernoulli is for a single trial, while Binomial is for multiple independent trials.
Bernoulli has one parameter (p), while Binomial has two parameters (n for the number of trials and
p for the probability of success in each trial).
Bernoulli gives the probability of a single success or failure, whereas Binomial gives the probability of
a specific number of successes in a fixed number of trials.
```

```python
Q7: Explain uniform Distribution with an example?
Ans:
    Uniform Distribution:
The uniform distribution is a probability distribution where all outcomes within a specified range are equally likely.
It is characterized by two parameters: a minimum value (a) and a maximum value (b).
Every value within the range [a, b] has the same probability of occurring.

Example:
Imagine rolling a fair six-sided die. The outcome can be any number from 1 to 6.
In this case, the uniform distribution describes the probability of each outcome:
P(X = 1) = P(X = 2) = P(X = 3) = P(X = 4) = P(X = 5) = P(X = 6) = 1/6
All six outcomes are equally likely, so each has a probability of 1/6
```

```python
Q8: What is the z score? State the importance of the z score?
Ans:
    Z-score:

A Z-score quantifies how many standard deviations a data point is from the mean in a dataset.
It's calculated as (X - μ) / σ, where X is the data point, μ is the mean, and σ is the standard deviation.

Importance of Z-score:
Standardizes data for comparisons.
Identifies outliers.
Calculates probabilities in normal distributions.
Facilitates data transformation and normalization for analysis
```

```python
Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem
Ans:
    Central Limit Theorem (CLT):
The Central Limit Theorem is a fundamental concept in statistics that states that when you take a large enough random
sample from any population with any shape of distribution (whether it's normal, skewed, or anything else) and calculate
the sample means, those sample means will be approximately normally distributed.

Significance of the Central Limit Theorem:
Normal Approximation: The CLT allows statisticians to approximate the sampling distribution of the sample mean with a normal
distribution, regardless of the underlying population distribution. This simplifies statistical analysis.

Inferential Statistics: It forms the basis for many statistical techniques and hypothesis tests. For example,
it enables the use of t-tests, ANOVA, and confidence intervals even when the population distribution is unknown.

Sample Size Determination: The CLT is used to determine the minimum sample size required for valid statistical inferences.
It ensures that sample means are normally distributed and can be used for hypothesis testing.

Quality Control: In industries like manufacturing and healthcare, the CLT is used to monitor and control processes by
analyzing sample means to make decisions about product quality or patient outcomes.

Data Modeling: When dealing with real-world data, which often doesn't follow a perfect normal distribution, the CLT allows
analysts to apply normal distribution-based models to approximate behavior.
```

```python
Q10: State the assumptions of the Central Limit Theorem?
Ans:
Random Sampling: The data must be collected through random sampling or a random process to ensure that the sample is 
representative of the population.

Independence: The individual observations in the sample must be independent of each other. The outcome of one observation
should not influence the outcome of another.

Sample Size: For the CLT to work effectively, the sample size should be sufficiently large. While there is no strict threshold,
a common guideline is that a sample size of at least 30 is often considered large enough for the CLT to apply.
```
