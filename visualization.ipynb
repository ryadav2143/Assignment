```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib?
Ans:
Matplotlib is a widely used Python library for creating static, interactive, and animated visualizations in a variety of formats and contexts.
It provides a wide range of tools for generating different types of plots, charts, and graphs to help visualize data in a more understandable 
and meaningful way. Matplotlib is particularly popular in the fields of data analysis, scientific research, and engineering due to its versatility
and ease of use.

The Pyplot module within Matplotlib provides a simple interface for creating various types of plots.
Here are five types of plots that can be created using the Pyplot module of Matplotlib:

Line Plot: Line plots are used to visualize the relationship between two variables by connecting data points with lines. They are commonly used
for time series data and to show trends over time.

Bar Plot: Bar plots display data using rectangular bars, with the length of each bar proportional to the value of the data it represents. They are
often used to compare different categories or groups.

Histogram: Histograms are used to represent the distribution of a dataset. They group data into intervals (bins) and display the frequency or count
of data points within each interval.

Scatter Plot: Scatter plots are used to show the relationship between two variables. Each data point is represented as a point on the graph, allowing
us to visualize patterns and correlations.

Pie Chart: Pie charts represent data as slices of a circular pie, where each slice's size is proportional to the data it represents. They are 
commonly used to show the composition of a whole in terms of its parts.
```

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot?
Ans:
A scatter plot is a type of data visualization that uses points to represent the values obtained for two different variables in a dataset. 
Each point on the scatter plot corresponds to a single observation or data point, and its position on the graph is determined by the values of
the two variables being plotted.

The primary purpose of a scatter plot is to show the relationship or correlation between the two variables. By plotting data points on a Cartesian
coordinate system (with one variable on the x-axis and the other on the y-axis), scatter plots allow you to visually identify patterns, trends, 
clusters, and outliers within the data.

import matplotlib.pyplot as mpl
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
mpl.scatter(x,y)
mpl.xlabel("x-axis")
mpl.ylabel("y-axis")
mpl.title("Scatter Graph")
mpl.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data?
Ans:
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. This is particularly useful when you want to display
multiple plots or visualizations in the same figure window, allowing you to compare and analyze different aspects of your data simultaneously. Each
subplot is a smaller plotting area within the larger figure, and you can arrange these subplots in rows and columns to organize your visualizations
efficiently.

import numpy as np
import matplotlib.pyplot as plt
a = np.array([0, 1, 2, 3, 4, 5])

x1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # Top-left subplot
plt.plot(y, x1, label='Line 1')
plt.legend()

plt.subplot(2, 2, 2)  # Top-right subplot
plt.plot(y, x2, label='Line 2')
plt.legend()

plt.subplot(2, 2, 3)  # Bottom-left subplot
plt.plot(y, x3, label='Line 3')
plt.legend()

plt.subplot(2, 2, 4)  # Bottom-right subplot
plt.plot(y, x4, label='Line 4')
plt.legend()

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot?
Ans:
A bar plot, also known as a bar chart, is a type of data visualization that uses rectangular bars to represent and compare the values of different
categories or groups. Each bar's length is proportional to the value it represents, making it easy to visually compare the magnitudes of these values.
Bar plots are used for various purposes:

Comparison: Bar plots allow for a quick visual comparison of values between different categories. They make it easy to identify which categories 
have larger or smaller values.

Distribution: Bar plots can show the distribution of data across categories. For example, you can create a bar plot to display the frequency of 
items falling into different ranges or bins.

Ranking: Bar plots can be used to rank categories based on their values. This is particularly useful when you want to show which categories are 
the most significant or prominent.

Change over Time: When using a time-based category, you can create grouped bar plots to visualize how values change over different time periods.

Comparison of Multiple Groups: Grouped bar plots can compare values across multiple sub-groups within each category, providing insights into 
different factors that contribute to the overall values.

Comparison of Multiple Variables: Stacked bar plots are useful for comparing the composition of multiple variables within each category.

import matplotlib.pyplot as mpl
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
mpl.bar(company,profit)
mpl.show
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot?
Ans:
A box plot, also known as a box-and-whisker plot, is a statistical data visualization that provides a concise summary of the distribution of a
dataset. It displays information about the median, quartiles, and potential outliers in the data. A box plot is particularly useful for identifying
the spread, skewness, and presence of outliers in a dataset.

Box plots are used for several purposes:

Identifying Distribution: Box plots provide an overview of the distribution of data, showing whether it is symmetric or skewed and whether
it has any extreme values.

Comparing Distributions: Box plots can be used to compare the distributions of different datasets or different groups within a dataset. This 
helps in identifying differences in central tendencies, spreads, and outliers.

Outlier Detection: Box plots can highlight potential outliers in the data, aiding in identifying values that are significantly different from 
the rest of the dataset.

Visualization of Summary Statistics: Box plots present key summary statistics (median, quartiles) in a concise and intuitive manner, making it
easier to understand the data's characteristics.

Assessing Skewness: The length and position of the box and the presence of outliers provide insights into the skewness of the data distribution.

import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.xlabel('Boxes')
plt.ylabel('Values')
plt.show()
```
