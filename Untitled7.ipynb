```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot?
Ans:
Seaborn is a powerful Python data visualization library based on Matplotlib. It provides a high-level interface for creating a variety 
of statistical graphics. Here are five types of plots that you can create using Seaborn, along with their uses:

Scatter Plot:
Use: Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents an
individual data point.

Bar Plot:
Use: Bar plots are commonly used to compare the values of different categories. They are useful for showing categorical data and their
corresponding values.

Histogram:
Use: Histograms are used to display the distribution of a single numerical variable. They help you understand the frequency or count
of data points within different ranges.

Line Plot:
Use: Line plots are often used to display data points connected by lines, which is useful for showing trends over time or some other
continuous variable.

Box Plot (Box and Whisker Plot):
Use: Box plots are great for visualizing the distribution of data, showing the median, quartiles, and potential outliers. They are 
particularly useful for identifying skewness and outliers in your data.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions?
Ans:
import seaborn as sns
fmri =sns.load_dataset('fmri')
sns.lineplot(x=fmri.timepoint , y= fmri.signal , data = fmri,hue = 'event',style = 'region')
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset?
Ans:
import seaborn as sns
tit = sns.load_dataset('titanic')
new_tit = sns.boxplot(x=tit['pclass'],y = tit['age'])
sns.jointplot(x=tit['fare'],y = tit.age)
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset?
Ans:
import seaborn as sns
dia = sns.load_dataset('diamonds')
sns.histplot(x = dia['price'] ,data = 'dia', hue = dia['cut'])
```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset?
Ans:
import seaborn as sns 
iris = sns.load_dataset("iris")
sns.pairplot(iris,hue = 'species')
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap?
Ans:
import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")
# Create a heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
# Set plot title and labels
plt.title("Passenger Count by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
