```python
Q2.What is the supervised learning ? List some Example of supervised learning?
Ans:
    Supervised learning is a type of machine learning in which an algorithm learns from labeled training data to make predictions
    or decisions. In supervised learning, the algorithm is provided with a dataset where each example in the dataset consists of 
    input data and the corresponding correct output or target. The algorithm's goal is to learn a mapping function that can 
    predict the correct output for new, unseen input data.

Examples of supervised learning:

Image Classification: Given a dataset of images with labels (e.g., pictures of cats and dogs), a supervised learning algorithm can
learn to classify new images into the correct categories.

Spam Email Detection: In this application, emails are labeled as either spam or not spam (ham). A supervised learning algorithm
can be trained to identify spam emails based on features of the email content.

Handwriting Recognition: An algorithm can be trained to recognize handwritten characters or digits by providing a dataset of handwritten
characters along with their corresponding labels.
```

```python
Q3- What is unsupervised learning? List some Examples of unsupervised learning?
Ans:
    Unsupervised learning is a type of machine learning where the algorithm learns from unlabeled data to discover patterns,
    structures, or relationships within the data. 
    
Examples of unsupervised learning include:

Clustering: Grouping similar data points together, such as K-Means Clustering.
Dimensionality Reduction: Reducing the number of features while preserving essential information, like Principal Component 
Analysis (PCA).
Anomaly Detection: Identifying unusual or outlier data points.
Topic Modeling: Discovering hidden topics within text data using techniques like Latent Dirichlet Allocation (LDA).
Recommendation Systems: Suggesting items or content to users based on their preferences and behavior.
Density Estimation: Estimating the underlying probability distribution of data points
```

```python
Q5- What are the main difference between supervised, unsupervised, and semi-supervised learning?
Ans:
Supervised Learning:
Uses labeled training data, where each input is paired with a corresponding output or target.
The algorithm learns to make predictions or classifications based on this labeled data.
Suited for tasks where you want the model to learn a mapping from inputs to known outputs.

Unsupervised Learning:
Uses unlabeled data, and the algorithm discovers patterns, structures, or relationships within the data.
There are no specific output labels or targets for the algorithm to predict.
Common applications include clustering and dimensionality reduction.

Semi-Supervised Learning:
Combines elements of both supervised and unsupervised learning.
Typically involves a small amount of labeled data and a larger amount of unlabeled data.
The model leverages the labeled data for guidance and the unlabeled data for additional learning or generalization.
Useful when acquiring labeled data is expensive or time-consuming, as it can provide improved performance compared to
unsupervised learning with limited labeled data.
```

```python
Q6- What is train, test and validation split? Explain the importance of each term?
Ans:
Training Set:
The training set is a portion of the dataset (typically the majority, around 70-80%) used to train the machine learning model.
The model learns from the patterns, relationships, and structures in the training data, updating its internal parameters to 
minimize prediction errors.
Importance: The training set is crucial because it is where the model learns and adapts to the underlying data. A well-trained
model should accurately capture the patterns present in the training data.

Test Set:
The test set is a separate portion of the dataset (usually 20-30%) that is not seen by the model during training.
It is used to evaluate the model's performance and assess how well it generalizes to new, unseen data.
Importance: The test set provides an objective measure of how well the model performs on data it hasn't encountered before.
It helps identify potential overfitting (model memorizing the training data) and provides an estimate of real-world performance.

Validation Set:
The validation set is an optional subset (typically 10-20%) used for model selection and hyperparameter tuning.
It helps assess different model variations and hyperparameter settings without contaminating the test set.
Importance: The validation set is crucial for optimizing a model's performance by fine-tuning parameters and selecting the best
model architecture. It prevents "data leakage" by ensuring that tuning decisions are based on data not used for testing.

The importance of these splits can be summarized as follows:
Training Set: Essential for model learning and adaptation.
Test Set: Essential for evaluating a model's performance on new, unseen data and detecting overfitting.
Validation Set: Essential for optimizing model parameters and selecting the best model configuration while preserving the integrity
of the test set for final evaluation.
```

```python
Q8- List down some commonly used supervised learning algorithms and unsupervised learning algprithms?
Ans:
Supervised Learning Algorithms:
Linear Regression: Used for regression tasks, predicting a continuous target variable.
Logistic Regression: Applied to binary classification problems, predicting one of two classes.
Decision Trees: Tree-like structures for both classification and regression tasks.
Random Forest: An ensemble method based on decision trees, offering improved accuracy.
Support Vector Machines (SVM): Useful for classification and regression tasks, particularly for complex decision boundaries.
K-Nearest Neighbors (KNN): A simple classification and regression algorithm based on the proximity of data points.

Unsupervised Learning Algorithms:
K-Means Clustering: Groups data points into clusters based on similarity.
Hierarchical Clustering: Builds a tree-like structure of nested clusters.
Principal Component Analysis (PCA): Reduces data dimensionality while preserving variance.
t-Distributed Stochastic Neighbor Embedding (t-SNE): Dimensionality reduction technique for visualization.
Gaussian Mixture Models (GMM): Probability-based clustering technique.
DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Clustering based on density of data points.
Anomaly Detection Algorithms: Include Isolation Forest and One-Class SVM for identifying outliers.
```
